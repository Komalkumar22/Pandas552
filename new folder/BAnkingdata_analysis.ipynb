```python
#"Calculate correlations between numerical columns. Which numerical features are most correlated with the target variable 'y'? Visualize this using a heatmap."

correlation_matrix = df.corr(numeric_only=True)

print(correlation_matrix)
```

```python

 #Convert target to numeric
df['y_numeric'] = (df['y'] == 'yes').astype(int)

# Select numeric columns for correlation
numeric_cols = ['age', 'balance', 'duration', 'campaign', 'pdays', 'previous', 'y_numeric']
correlation_matrix = df[numeric_cols].corr()

print("\nCorrelation with Campaign Response (y):")
response_correlation = correlation_matrix['y_numeric'].sort_values(ascending=False)
print(response_correlation)
```

```python
import seaborn as sns
plt.figure(figsize=(5,5))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, 
            square=True, linewidths=1)
plt.title('Correlation Matrix - Key Features')
plt.tight_layout()
plt.savefig('correlation_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Create segments based on age groups (18-30, 30-45, 45-60, 60+). For each segment, calculate subscription rate and average balance. Which age group has the highest conversion?
# Create age groups
bins = [18, 30, 45, 60, float('inf')]
labels = ['18-30', '30-45', '45-60', '60+']

df['Age_Group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
```

```python
# Convert target variable to numeric
df['y_numeric'] = df['y'].map({'yes': 1, 'no': 0})


age_summary = (
    df.groupby('Age_Group')
      .agg(
          Subscription_Rate=('y_numeric', 'mean'),
          Average_Balance=('balance', 'mean'),
          Total_Customers=('Age_Group', 'count')
      )
)
```

```python
age_summary
```

```python
#Create new features: (a) 'age_group' from age, (b) 'high_balance' (True if balance > median), (c) 'campaign_category' (Low: 1-3, Medium: 4-6, High: 7+). How does subscription rate vary across these new features?
df['age_group']=df['Age_Group']
```

```python
# high_balance = 
# (for b in df['balance']
#     if b> df['balance'].median()
#         print('True')
#     else 'False'
# )

high_balance = [
    'True' if b > df['balance'].median() else 'False' 
    for b in df['balance']
]
```

```python
df['high_balance'] = df['balance'] > df['balance'].median()
```

```python
# Define the boundaries and the labels for your categories
bins = [0, 3, 6, float('inf')]
labels = ['Low', 'Medium', 'High']

# Create the new column
df['campaign_category'] = pd.cut(df['campaign'], bins=bins, labels=labels)
```

```python
age_summary = (
    df.groupby('age_group')['y_numeric']
      .mean()
      .mul(100)
      .round(2)
)

print(age_summary)
```

```python
balance_summary = (
    df.groupby('high_balance')['y_numeric']
      .mean()
      .mul(100)
      .round(2)
)

print(balance_summary)
```
